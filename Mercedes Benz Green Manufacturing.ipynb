```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
```

```python
train_data = pd.read_csv("BENZ_train.csv")
test_data = pd.read_csv("BENZ_test.csv")
```

# checking for NULL Values

```python
for i in range(len(train_data.columns)):
    if train_data.iloc[:,i].isnull().sum() == 0 :
        continue
    else:
        print(i,"th column has null values")
```

```python
for i in range(len(test_data.columns)):
    if test_data.iloc[:,i].isnull().sum() == 0 :
        continue
    else:
        print(i,"th column has null values")
```

# therefore both train data and test data has no NULL Values

# label encoding

```python
le = LabelEncoder()
```

```python
for i in train_data.columns:
    if train_data[i].dtype == "object":
        train_data[i] = le.fit_transform(train_data[i])
```

```python
train_data
```

```python
for i in test_data.columns:
    if test_data[i].dtype == "object":
        test_data[i] = le.fit_transform(test_data[i])
```

```python
test_data
```

# removing unneccesary columns
#columns with low variance

```python
for i in train_data.columns[2:]:
    if train_data[i].var() < 0.01 and test_data[i].var() < 0.01 :
        train_data.drop(i,axis = 1,inplace = True)
        test_data.drop(i,axis = 1,inplace = True)
    else:
        continue
```

# TRAIN & TEST DATA

```python
x_train = train_data.drop(["ID","y"],axis =1)
y_train = train_data["y"]
x_test = test_data.drop(["ID"],axis=1)
```

```python
print("shape of x_train",x_train.shape)
print("shape of y_train",y_train.shape)
print("shape of x_test",x_test.shape)
```

# DIMENSIONALITY REDUCTION

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=30)
```

```python
pca.fit(x_train)
x_train_pca =  pca.transform(x_train)
print("BEFORE PCA",x_train.shape)
print("after PCA",x_train_pca.shape)
```

```python
pca.fit(x_test)
x_test_pca =  pca.transform(x_test)
print("BEFORe PCA",x_test.shape)
print("after PCA",x_test_pca.shape)
```

# MODEL FITTING XGBOOSTING

```python
from xgboost import XGBRegressor
model = XGBRegressor(n_estimators=1000, max_depth=7, eta=0.01, subsample=0.7, colsample_bytree=0.8)
```

```python
from sklearn.model_selection import KFold,cross_val_score
k_fold = KFold(n_splits=10, shuffle=True, random_state=7)
```

```python
scores = cross_val_score(model, x_train_pca, y_train, scoring='neg_mean_absolute_error', cv=k_fold, n_jobs=-1)
```

```python
scores_ = abs(scores)
print("average MAE :%.3f (%.3f)"%(scores_.mean(),scores_.std()))
```

# Predicting for TEST DATA

```python
model.fit(x_train_pca,y_train)
y_pred = model.predict(x_test_pca)
```
